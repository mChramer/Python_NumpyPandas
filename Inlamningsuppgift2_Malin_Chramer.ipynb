# Part 1 - Numpy

```python
import numpy as np
```

#### 1. Konstruera en array med 10 kolumner och 5 rader av float-tal. Inget tal får vara samma.

```python
Array1 = np.arange(1.0,51.0).reshape(5,10)
print(Array1)
```

#### 2. Konstruera en ny array till med 10 kolumner och 7 rader av float-tal. Den skall inte vara samma som ovan array.

```python
Array2 = np.arange(51.0,121.0).reshape(7,10)
print(Array2)
```

#### 3. . Gör om de två arrayerna från föregående uppgifter så att de har 5 kolumner, med så många extra rader som behövs.

```python
Array1 = Array1.reshape(10,5)
print(Array1)
```

```python
Array2 = Array2.reshape(14,5)
print(Array2)
```

#### 4. Ta nu de två arrayerna från de föregående uppgifterna och stapla ihop dem till en array som fortfarande har fem kolumner.

```python
# alternativ 1
Array3 = np.vstack([Array1, Array2])
print(Array3)
```

```python
Array3.ndim
```

```python
# alternativ 2
Array4 = np.concatenate((Array1, Array2))
print(Array4)
```

```python
Array4.ndim
```

#### 5. Ändra ordningen på kolumnerna så att de har ordningen kolumn5, kolumn3, kolumn4, kolumn1 och kolumn2.

```python
i = [4,3,2,0,1]
swapped_columns = Array4[:,i]
```

```python
print(swapped_columns)
```

#### Step 4 - fråga och svara på 4 frågor om din data.

Mina frågor:

    1. Vilka är de vanligaste orsakerna till att en hund kommer till ett hundhem?
    
    2. Vilka raser är de vanligaste?
    
    3. Hur är det vanligaste sättet en hund får nytt hem?
    
    4. Är det flest hanar eller tikar som är på hundhemmen?

```python
#  1. Vilka är de vanligaste orsakerna till att en hund kommer till ett hundhem?

intakereason = df(dogs_data.intakereason.value_counts())
intakereason.plot.bar()
```

Den vanligaste orsaken till att en hund hamnar på hundhem är att den är herrelös, datan som jag har ger ingen förklaring till varför hunden är herrelös men det vanligaste är att hunden blivit övergiven av sina tidigare ägare.
På andra plats är att hunden inte längre passar in i familjens/ägaren liv och att man då väljer att lämna sin hund till ett hundhem där den kan få en ny ägare.
De sju följande orsakerna ligger sedan på ganska jämna siffror.

```python
# 2. Vilka raser är de vanligaste?
# - hur stor procentandel är varje ras i förhållande till totala antalet hundar?
```

```python
breed = df(dogs_data.breedname.value_counts()) # räknar hur många som finns av varje ras.
breed = breed.head(5) # visar de fem största värdena.
breed
```

```python
breed["%"] = breed["breedname"] / breed["breedname"].sum() *100 # räknar ut procentandel
breed.plot.pie(y="%", figsize=(5,5), autopct='%1.2f%%', legend=None) # väljer diagram, tar bort legend
plt.title(label="5 most common breeds in percent", fontdict={"fontsize":16}, pad=20) # anger rubrik
plt.ylabel("") # ger ylabel en tom string så den ej syns.
plt.show
```

Av det totala antalet hundar på hundhem kan vi här se att den största delen av raseer är 'Bully breed mix' och 'Labrador Retriver'. Tillsammans utgör dessa nästan 50% av totala antalet hundar.

```python
# 3. Hur är det vanligaste sättet en hund får nytt hem?
# - hur många procent jämfört med totala antalet hundar?
```

```python
move = df(dogs_data.movementtype.value_counts())
move
```

```python
move["movementtype"] = move["movementtype"] / breed["breedname"].sum() *100 # räknar ut procentandelen
move.plot.pie(y="movementtype", figsize=(5,5), autopct='%1.2f%%', legend=None) # väljer vilket diagram, tar bort legend
plt.title(label="Reason for new home", fontdict={"fontsize":16}, pad=20) # sätter rubrik på diagrammet
plt.ylabel("") # sätter y till en tom string så den ej syns
plt.show
```

Av alla hundar är det ca 56% som blir adopterade av nya familjer/ägare. Samt ca 30% kommer tillbaka till sina ägare.
